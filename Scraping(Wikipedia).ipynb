```python
r = requests.get(url)

soup = BeautifulSoup(r.content, 'html.parser')
#print()
```

```python
print(r)
```

```python
print(soup.prettify())
```

```python
table = soup.find('div', class_="mw-parser-output")
```

```python
print(table.text)
```

```python
lists = table.find_all('li')
for list in lists:
    #print(list,end="\n"*2)
    link = list.find("a")
    link_url=link["href"]
    print(f"https://en.wikipedia.org/{link_url}")
    


    #print(f"Apply here: {link_url}\n")
```
