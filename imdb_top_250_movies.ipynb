```python
import requests
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36'}
url = 'https://www.imdb.com/chart/top/'
r = requests.get(url,headers = headers).text
```

```python
soup = BeautifulSoup(r, 'lxml')
```

```python
list = soup.find('ul', class_='ipc-metadata-list')
for l in list:
    print(l.text.strip())
    print()
```

```python
each = list.find_all('li', class_="ipc-metadata-list-summary-item")
```

```python
title = []
year_of_release = []
movie_duration = []
movie_ratings = []

rating = []

for i in each:
    title.append(i.find('h3', class_='ipc-title__text').text)
    #print(i.find_all('span', class_='sc-14dd939d-6')[2])
    release = i.find_all('span', class_='sc-14dd939d-6')[0]
    duration = i.find_all('span', class_='sc-14dd939d-6')[1]
    try:
        movie_rating = i.find_all('span', class_='sc-14dd939d-6')[2]
    except IndexError:
        movie_rating = None
    
    for j in release:
        year_of_release.append(j.text)
       
    for j in duration:
            movie_duration.append((j.text))
            
    if movie_rating is not None:
        for j in movie_rating:
            movie_ratings.append((j.text))
    else:
        movie_ratings.append("none")
            
    rating.append(i.find('span', class_='ipc-rating-star').text)
    #print(i.find('span', class_='sc-14dd939d-6').text)
    #print(i.find('span', class_='sc-14dd939d-6').text)
    #print()
```

```python
d = {'Movie-Title': title,'Year_of_Release':year_of_release,'Rating':movie_duration, 'movie_rating': movie_ratings, 'Review_rating': rating}
df = pd.DataFrame(d)
    
```
