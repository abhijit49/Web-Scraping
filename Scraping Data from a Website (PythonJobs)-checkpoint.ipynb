# Extracting all the Tags

```python
for job_element in job_elements:
    link = job_element.find("a", class_="go_button")
    link_url = link["href"]
    

    new_link = f"https://pythonjobs.github.io{link_url}"
    #print(new_link)

    new_response = requests.get(new_link)
    new_soup = BeautifulSoup(new_response.content,'html.parser')

    job_element = new_soup.find("h1")
    print("Title:",job_element.text.strip())
    
    tag_elements = new_soup.select(".head .tags a.tag")
    print("Tags:",end="")
    for tag_element in tag_elements:
        all_tag = tag_element.get_text()
        print("\t",all_tag.strip())
    print()     
 
    
print()
```

# Extracting Contacts

```python
for job_element in job_elements:
    link = job_element.find("a", class_="go_button")
    link_url = link["href"]
    

    new_link = f"https://pythonjobs.github.io{link_url}"
    #print(new_link)

    new_response = requests.get(new_link)
    new_soup = BeautifulSoup(new_response.content,'html.parser')

    job_element = new_soup.find("h1")
    print("Title:",job_element.text.strip())
    
    contacts = new_soup.find("div", class_="contact").find_all("div", class_="field")
    for con in contacts:
        contact = con.get_text().strip()
        print(contact)
    
    print()
```
